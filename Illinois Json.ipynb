```python
import requests
import pandas as pd
import matplotlib as plt
import json
```

```python
response = requests.get('https://www.dph.illinois.gov/sitefiles/COVIDZip.json?nocache=1')

data = response.json()
Zip_Df= pd.json_normalize(data['zip_values'])
```

```python
Zip_Df
```

```python
# Create DataFrame of Zip and Age
Zip_Age = Zip_Df.drop(['confirmed_cases', 'demographics.race', 'demographics.gender'], axis=1)

Zip_Age.set_index('zip', inplace=True)
for index, row in Zip_Age.iterrows():
    ages = row['demographics.age']
    for age_group in ages:
        Zip_Age.at[index, age_group['age_group']] = age_group['count']

Zip_Age= Zip_Age.drop(columns='demographics.age') 
Zip_Age
```

```python
Zip_Race = Zip_Df.drop(['confirmed_cases', 'demographics.age', 'demographics.gender'], axis=1)
Zip_Race.set_index('zip', inplace=True)
for index, row in Zip_Race.iterrows():
    races = row['demographics.race']
    for race_group in races:
        Zip_Race.at[index, race_group['description']] = race_group['count']
Zip_Race= Zip_Race.drop(columns='demographics.race') 
Zip_Race
```

```python
Zip_Gender = Zip_Df.drop(['confirmed_cases', 'demographics.age', 'demographics.race'], axis=1)
Zip_Gender.set_index('zip', inplace=True)
for index, row in Zip_Gender.iterrows():
    genders = row['demographics.gender']
    for gender_group in genders:
        Zip_Gender.at[index, gender_group['description']] = gender_group['count']
Zip_Gender= Zip_Gender.drop(columns='demographics.gender') 
Zip_Gender
```

```python
Zip_Gender.to_csv('Illinois_Gender.csv')
Zip_Race.to_csv('Illinois_Race.csv')
Zip_Age.to_csv('Illinois_Age.csv')
```
