```python
#Dependencies
import json
import os
import pandas as pd
from pprint import pprint
from pandas.io.json import json_normalize
```

```python
# Load JSON
filepath = os.path.join('./Input Data/Illinois_Covid_Json')
with open(filepath) as jsonfile:
    covid_json = json.load(jsonfile)

pprint(covid_json)
```

```python
# Create DataFrame by zip
Zip_Df = json_normalize(covid_json["zip_values"])
Zip_Df = pd.DataFrame(Zip_Df)
Zip_Df
```

```python
# Create DataFrame of Zip and Confirmed Cases
Zip_Confirmed = Zip_Df.drop(['demographics.age', 'demographics.race', 'demographics.gender'], axis=1)
Zip_Confirmed

#Output to csv
Zip_Confirmed.to_csv('./Data Output/Zip_Confirmed.csv', index = False, header=True)
```

```python
# Create DataFrame of Zip and Age  
Zip_Age = Zip_Df.drop(['confirmed_cases', 'demographics.race', 'demographics.gender'], axis=1)

Zip_Age.set_index("zip", inplace=True)
for index, row in Zip_Age.iterrows():
    ages = row["demographics.age"]
    for age_group in ages:
        Zip_Age.at[index, age_group["age_group"]] = age_group["count"]
        
Zip_Age = Zip_Age.drop(['demographics.age'], axis=1)
Zip_Age

#Output to csv
Zip_Age.to_csv('./Data Output/Zip_Age.csv', index = True, header=True)
```

```python
# Create DataFrame of Zip and Race  
Zip_Race = Zip_Df.drop(['confirmed_cases', 'demographics.age', 'demographics.gender'], axis=1)
Zip_Race

Zip_Race.set_index("zip", inplace=True)
for index, row in Zip_Race.iterrows():
    races = row["demographics.race"]
    for description in races:
        Zip_Race.at[index, description["description"]] = description["count"]
        
Zip_Race = Zip_Race.drop(['demographics.race'], axis=1)
Zip_Race

#Output to csv
Zip_Race.to_csv('./Data Output/Zip_Race.csv', index = True, header=True)
```

```python
# Create DataFrame of Zip and Gender  
Zip_Gender = Zip_Df.drop(['confirmed_cases', 'demographics.age', 'demographics.race'], axis=1)
Zip_Gender

Zip_Gender.set_index("zip", inplace=True)
for index, row in Zip_Gender.iterrows():
    gender = row["demographics.gender"]
    for description in gender:
        Zip_Gender.at[index, description["description"]] = description["count"]
        
Zip_Gender = Zip_Gender.drop(['demographics.gender'], axis=1)
Zip_Gender

#Output to csv
Zip_Gender.to_csv('./Data Output/Zip_Gender.csv', index = True, header=True)
```
