```python
import pandas as pd
import os, sys
import glob
import numpy as np
```

```python
path = "csse_covid_19_data/csse_covid_19_daily_reports"
all_files = glob.glob(path + "/*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, encoding="ISO-8859-1")
    li.append(df)
    frame = pd.concat(li, axis=0, ignore_index=True, sort=False)
```

```python
frame.count()
```

```python
df = frame
df = df.fillna('')
df['Province_State'] = df['Province/State'].astype(str) + df['Province_State'].astype(str)
df['Country_Region'] = df['Country/Region'].astype(str) + df['Country_Region'].astype(str)
df['Last_Update'] = df['Last Update'].astype(str) + df['Last_Update'].astype(str)
df['Latitude'] = df['Latitude'].astype(str) + df['Lat'].astype(str)
df['Longitude'] = df['Longitude'].astype(str) + df['Long_'].astype(str)
df = df.drop(columns=['Province/State','Country/Region','Last Update','Lat','Long_','Combined_Key'])
df['Last_Update'] = df['Last_Update'].str[0:10]
df = df.sort_values(by=['Country_Region','Last_Update'])
df.drop_duplicates(keep='first', inplace=True)
df = df[['Province_State','Country_Region','Confirmed','Deaths','Recovered','Active','Latitude','Longitude','FIPS','Admin2']]
df
```

```python
df.to_csv("clean_data.csv")
```
